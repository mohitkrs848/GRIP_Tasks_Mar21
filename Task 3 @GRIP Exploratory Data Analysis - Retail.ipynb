### Covariance of Dataset
Covariance is a measure of how much two random variables vary together. It’s similar to variance, but where variance tells you how a single variable varies, co variance tells you how two variables vary together.
Cov(X,Y) = Σ E((X-μ)E(Y-ν)) / n-1 where:
X is a random variable
E(X) = μ is the expected value (the mean) of the random variable X and
E(Y) = ν is the expected value (the mean) of the random variable Y
n = the number of items in the data set

```python
dataset.cov()
```

```python
#heatmap of Covarriance
axes = plt.subplots(1,1,figsize=(10,6))
sns.heatmap(dataset.cov(),annot = True)
plt.show()
```

#### Bar chart or Bar graph 
is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

```python
plt.figure(figsize=(18,6))
plt.bar('Sub-Category','Category',data = dataset)
plt.title('Sub-Categories in different Categories',size = 30 ,pad = 10)
plt.xlabel('Sub-Categories ->' ,size = 25)
plt.xticks(rotation=45)
plt.ylabel('Categories ->',size = 25)
plt.show()
```

```python
#A scatter plot is a diagram where each value in the data set is represented by a dot.
#The Matplotlib module has a method for drawing scatter plots, it needs two arrays of the same length, 
#one for the values of the x-axis, and one for the values of the y-axis
dataset.plot.scatter(x='Discount',y='Profit',color='green',marker='.',alpha=0.4)
```

###### Business Insights : answer us in understanding the relationship between discount and profit ,that from the graph we get to kow that when there is less discount we have more profit but when we gives high discount then we got less profit.

#### hist()
function is defined as a quick way to understand the distribution of certain numerical variables from the dataset. It divides the values within a numerical variable into "bins". It counts the number of examinations that fall into each of the bin. These bins are responsible for a rapid and intuitive sense of the distribution of the values within a variable by visualizing bins.

```python
dataset.hist(bins = 10,figsize = (20,15),facecolor = 'green',alpha =0.4)
plt.show()
```

```python
#pie-chart

plt.figure(figsize=(15,12))
dataset['Sub-Category'].value_counts().plot.pie()
plt.show()
```

###### Business Insights: shows us the sub-category products part in total sales.

#### countplot()
method is used to Show the counts of observations in each categorical bin using bars.

```python
plt.figure(figsize=(15,12))
sns.countplot(x='State',data=dataset,palette='spring_r',order=dataset['State'].value_counts().index)
plt.xticks(rotation=60)
plt.show()
```

###### Business Insight : Helps in getting that which state have the highest selling products of our company and where we need to do more marketing ,so that everybody in that region is going to know about our company .
